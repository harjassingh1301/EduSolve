```python
from modelscope.pipelines import pipeline
from modelscope.outputs import OutputKeys

p = pipeline('text-to-video-synthesis','damo/text-to-video-synthesis')
```

```python
from googletrans import Translator
from gtts import gTTS
import os
from IPython.display import Audio
#from playsound import playsound

langdest=""
menu='''Language menu:
    1. Hindi
    2. Bengali
    3. Telugu
    5. Marathi
    6. Tamil
    7. Urdu
    8. Gujurati
    9. Nepali'''
texti=input("Enter the text to be translated here: ")
langtr=input(" Choose a language to which you want your text translated to from the given menu: ")
langs=[["Hindi","hi"],["Bengali","bn"],["Telugu","te"],["Marathi","mr"],["Tamil","ta"],["Urdu","ur"],["Gujurati","gu"],["Nepali","ne"]]
for i in langs:
    if langtr==i[0]:
        langdest=i[1]
trans=Translator()
out= trans.translate(texti, dest=langdest)
a=out.text
print(a)
voice = gTTS(a, lang=langdest)
voice.save("voice.mp3")
Audio("voice.mp3", autoplay=True)
```

```python
test_text = {
    'text': texti
}
output_video_path = p(test_text,)[OutputKeys.OUTPUT_VIDEO]
print ('output_video_path', output_video_path)
```
